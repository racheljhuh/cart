### Step 2. Plot the data in Python

````python
# Do this first! these are the libraries we need
import numpy as np
import matplotlib.pyplot as plt
````

### Matplotlib documentation!
Link: https://matplotlib.org/stable/api/pyplot_summary.html

````python
# plot univariate box and whisker
plt.boxplot(price)
plt.title('Univariate box and whisker')
plt.ylabel('Price ($)')
````

````python
# #Create scatter plot
x = np.ones(len(price))

plt.scatter(x, price)
plt.title('Univariate Scatter Plot')
plt.ylabel('Price ($)')
plt.xlim((.95, 1.5))
````

### 2. Using 2 distinct bivariate data representations, plot price by region

Screenshot 2023-10-16 at 4.53.16 PM.png


````python
#we can stack plots, add error bars, connect means, etc
plt.bar(region, price, alpha = .5)
plt.scatter(region, price)
plt.title('Bivariate - Bar and Scatter')
plt.xlabel('Region')
plt.ylabel('Price ($)')
plt.savefig('bivariate.jpg')
plt.savefig('bivariate.svg')
````

### 3. Using 1 trivariate data representation, plot price, region, miles/gallon

Screenshot 2023-10-16 at 4.52.32 PM.png

Screenshot 2023-10-16 at 4.53.44 PM.png

````python
fig, axs = plt.subplots(nrows=1, ncols=3)
axs[0].scatter(price[0:3],mpg[0:3]) #limit to only EU cars
axs[0].set_title('EU cars')
axs[1].scatter(price[3:6],mpg[3:6]) #limit to only NA cars
axs[1].set_title('NA cars')
axs[2].scatter(price[6:10],mpg[6:10]) #limit to only AS cars
axs[2].set_title('AS cars')
````

4. Using 1 hypervariate data representation: plot price, region, miles/gallon, and if the price is for electric or gas vehicles of the producer.

````python
normprice = price/(np.max(price))
normprice = 1000*normprice**2
print(normprice)
plt.scatter(region, mpg, s=normprice, c=engine, alpha = 0.5)
plt.scatter(region,mpg, c='black')
plt.xlabel('Region')
plt.ylabel('MPG')
plt.title('Hypervariate Plot')
# plt.savefig('variate.svg')
````

### Oof... Those colors aren't great. We can do all sorts of editing in Python, but we can also move these over to Figma!

How? We can save our favorite plots as vector-based images (.SVG) and bring the images directly into our current Figma frame.

We want to avoid using pixel-based images (.JPG) because:
1. We can't edit elements in Figma
2. The resolution degrades rapidly
